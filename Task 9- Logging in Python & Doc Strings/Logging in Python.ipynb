```python
import logging


# basicConfig(filename='bank_info.log', filemode='a', level=INFO, style='{',
#             format=' {lineno} || {asctime} -- {name} -- {levelname} -- {message}')
info_logger = logging.getLogger('Pranay')
info_logger.setLevel(logging.INFO)
info_handler = logging.FileHandler("bank_info.log",)
info_Formatter = logging.Formatter(
    ' {lineno} || {asctime} -- {name} -- {levelname} -- {message}', style='{')
info_handler.setFormatter(info_Formatter)
info_logger.addHandler(info_handler)

debug_logger = logging.getLogger('PranayDebug')
debug_logger.setLevel(logging.DEBUG)
debug_handler = logging.FileHandler("bank_debug.log",)
debug_Formatter = logging.Formatter(
    ' {lineno} || {asctime} -- {name} -- {levelname} -- {message}', style='{')
debug_handler.setFormatter(debug_Formatter)
debug_logger.addHandler(debug_handler)

warning_logger = logging.getLogger('PranayWarning')
warning_logger.setLevel(logging.WARNING)
warning_handler = logging.FileHandler("bank_warning.log",)
warning_Formatter = logging.Formatter(
    ' {lineno} || {asctime} -- {name} -- {levelname} -- {message}', style='{')
warning_handler.setFormatter(warning_Formatter)
warning_logger.addHandler(warning_handler)

error_logger = logging.getLogger('PranayError')
error_logger.setLevel(logging.ERROR)
error_handler = logging.FileHandler("bank_error.log",)
error_Formatter = logging.Formatter(
    ' {lineno} || {asctime} -- {name} -- {levelname} -- {message}', style='{')
error_handler.setFormatter(error_Formatter)
error_logger.addHandler(error_handler)


class Bank_Account:
    database = []

    # Logging configurations for different log levels

    def __init__(self):
        info_logger.info("Bank System Started")
        self.account_number = None
        self.account_name = None
        self.balance = 0
        self.previousBalance = 0

    def createAccount(self):
        account_number = int(input("Enter Account Number: "))
        if account_number in self.database:
            print("Account Already Exists")
            self.login()
        else:
            print("You are not Registered!! Create Account")
            account_number = int(input("Enter Account Number: "))
            account_name = input("Enter Name: ")
            print("Account Created Successfully!!")
            self.account_number = account_number
            self.account_name = account_name
            self.balance = 0
            self.previousBalance = 0
            self.database.append({
                "account_number": self.account_number,
                "account_name": self.account_name,
                "balance": self.balance,
                "previous_balance": self.previousBalance
            })

    def login(self):
        self.account_number = int(input("Enter Account Number: "))
        if self.account_number in [account["account_number"] for account in self.database]:
            self.choose_option()
        else:
            print("Account Not Found")
            warning_logger.warning('Account not Exists')
            self.createAccount()

    def choose_option(self):
        opt = int(input("\nEnter Option\n 1. Deposit\n 2. Withdrawal\n"))
        if opt == 1:
            self.deposit()
        elif opt == 2:
            self.withdraw()

    def deposit(self):
        debug_logger.debug("Amount deposit Started")
        amount = float(input("Enter amount to be Deposited: "))
        self.balance += amount
        print("\n Amount Deposited Successfully! Amount:", amount)
        print("\n Your Previous Balance before deposit:", self.previousBalance)
        self.update_database()
        info_logger.info(f"Amount Deposited: {amount}")
        info_logger.info(f"Amount Deposited Successfully")

    def withdraw(self):
        debug_logger.debug("Amount Withdrawal Started")
        try:
            amount = float(input("Enter amount to be Withdrawn: "))
            if self.balance >= amount:
                self.balance -= amount
                print("\n You Withdrew amount:", amount)
                self.previousBalance = self.balance + amount
                self.update_database()
                debug_logger.debug("Amount Withdrawal Successful")
                info_logger.info(f"Amount Withdrawn: {amount}")
            else:
                print("\n Insufficient balance")
                warning_logger.warning(
                    "Insufficient balance, please enter a valid amount")
        except:
            print("please enter a valid amount")
            error_logger.error("Invalid Amount Entered !!")

    def update_database(self):
        for account in self.database:
            if account["account_number"] == self.account_number:
                account["balance"] = self.balance
                account["previous_balance"] = self.previousBalance

    def display(self):
        for account in self.database:
            if account["account_number"] == self.account_number:
                print("\n Account Number:", account["account_number"])
                print(" Account Name:", account["account_name"])
                print(" Net Available Balance:", account["balance"])
                print(" Previous Balance:", account["previous_balance"])


# creating an object of class
s = Bank_Account()
s.createAccount()
s.login()

s.choose_option()

s.display()


labels = ['ID', 'Name', 'Net Balance', 'Previous Balance']

for account in s.database:
    print("\n")
    for label, value in zip(labels, account.values()):
        print(f"{label}: {value}")

for account in s.database:
    print(account)  # to rough check database info
```
