#### Saving a Pandas Dataframe as csv

```python

import pandas as pd

name = ["aparna", "pankaj", "sudhir", "Geeku"]
degree = ["MBA", "BCA", "M.Tech", "MBA"]
score = [90, 40, 80, 98]
 
# dictionary of lists
dict = {'name': name, 'degree': degree, 'score': score}
     
df = pd.DataFrame(dict)
 
print(df)
```

```python
df.to_csv('file1.csv')
```

```python
df.to_csv('file2.csv', header=False, index=False)
```

### Loading Excel spreadsheet as pandas DataFrame

```python
excel_file = pd.ExcelFile('Salary1.xlsx')
  
# View the excel_file's sheet names
print(excel_file.sheet_names)
  
# Load the excel_file's Sheet1 as a dataframe
df = excel_file.parse('Salary_Data')
print(df)
```

```python
require_cols = [0, 3] 
    
# only read specific columns from an excel file 
required_df = pd.read_excel('Salary1.xlsx', usecols = require_cols) 
    
print(required_df) 
```
