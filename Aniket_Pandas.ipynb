## Data Cleaning
Data cleaning means fixing bad data in your data set.

Bad data could be:

Empty cells
Data in wrong format                                                                                                         Wrong data
Duplicates

#### Remove Rows
One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

```python
df
```

```python
new_df = df.dropna()

print(new_df.to_string())
```

If you want to change the original DataFrame, use the inplace = True argument:

```python
df.dropna(inplace = True)

print(df.to_string())
```

#### Replace Empty Values
Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

```python
df.fillna(130, inplace = True)
```

```python
df["Years of Experience"].fillna(13, inplace = True)
```

#### Replace Using Mean, Median, or Mode
A common way to replace empty cells, is to calculate the mean, median or mode value of the column.

Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column:

```python
x = df["Years of Experience"].mean()

df["Years of Experience"].fillna(x, inplace = True)
```

# concat()

We can perform concatenation of pandas object by axis wise as

By default, axis=0, i.e. row-wise concatenation,

so if we set axis=1, column-wise concatenation will be performed.

```python
df6 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df6
```

```python
df7 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4,5,6,7])
df7
```

```python
df_con1 = pd.concat([df6,df7], axis=0)
df_con1
```

```python
df_con2 = pd.concat([df6,df7], axis=1)
df_con2
```

```python
df_con2.fillna(0, inplace=True)
print(df_con2)
```

```python

import pandas as pd
import numpy as np

# create sample data for patient dataframe
np.random.seed(123)
patient_ids = ['P' + str(i) for i in range(1,26)]
names = ['Patient ' + str(i) for i in range(1,26)]
ages = np.random.randint(20, 80, size=25)
genders = np.random.choice(['Male', 'Female'], size=25)
admission_dates = pd.date_range(start='2022-01-01', end='2022-01-25')
discharge_dates = pd.date_range(start='2022-01-05', end='2022-01-29')
diagnoses = ['Fever', 'Flu', 'Malaria', 'Tuberculosis', 'Pneumonia', 'Bronchitis', 'Heart Disease']

# create patient dataframe
df = pd.DataFrame({
    'Patient ID': patient_ids,
    'Name': names,
    'Age': ages,
    'Gender': genders,
    'Admission Date': admission_dates,
    'Discharge Date': discharge_dates,
    'Diagnosis': np.random.choice(diagnoses, size=25)
})

print(df.head())
```
