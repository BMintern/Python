```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import json
import numpy as np
from datetime import datetime
```

```python
# get html data first
html_data = requests.get("https://en.wikipedia.org/wiki/World_Happiness_Report")

# check if status is 200 -> shows that its allowed to scrape the webpage
print(html_data.status_code)
```

```python
# parse html data now using BeautifulSoup
soup = BeautifulSoup(html_data.text, "html.parser")

# get all tables from wikipedia page
tables = soup.find_all('table',{'class':"wikitable"})

# store target table
table = tables[3]

# convert table html code to pandas df
data = pd.read_html(str(table))
df_happiness = pd.DataFrame(data[0]) 

# show head
df_happiness.head()
```

```python
# rename some countries to later match the country names from RapidAPI
df_happiness = df_happiness.apply(lambda x: x.replace("Congo (Kinshasa)", "DR Congo"))
df_happiness = df_happiness.apply(lambda x: x.replace("Congo (Brazzaville)", "Congo"))
df_happiness = df_happiness.apply(lambda x: x.replace("Ivory Coast", "Côte d'Ivoire"))
```

```python
# print number of countries in the dataset
list_countries = df_happiness["Country or region"].to_list()
print(f"Number of countries in Wikipedia dataset: {len(list_countries)}")
```

```python
from tqdm import tqdm
import requests
import json
import numpy as np
import pandas as pd

# Rename some countries to later match the country names from RapidAPI
df_happiness["Country or region"] = df_happiness["Country or region"].replace(
    {
        "Congo (Kinshasa)": "DR Congo",
        "Congo (Brazzaville)": "Congo",
        "Ivory Coast": "Côte d'Ivoire"
    }
)

# Create URL and headers for API call
url = "https://world-population.p.rapidapff1ba047e1msh847b15bfd7c32e6p16051bjsne6e319c0a978i.com/population"

# The headers can be found when logging in to your RapidAPI account and opening the link above
headers = {
    'x-rapidapi-host': "world-population.p.rapidapi.com",
    'x-rapidapi-key': "ff1ba047e1msh847b15bfd7c32e6p16051bjsne6e319c0a978"
}

# Create a copy of the dataframe to store the population data
df_happiness_with_population = df_happiness.copy()

# Loop over countries and get population
for idx, row in tqdm(df_happiness.iterrows(), total=df_happiness.shape[0], desc="Fetching population"):
    country = row["Country or region"]
    
    # Create querystring for API call
    querystring = {"country_name": country}
    
    # Create request and fetch response
    response = requests.request("GET", url, headers=headers, params=querystring)
    
    # Parse the response
    response_dict = json.loads(response.text)
    print(response_dict)
    
    if "ok" in response_dict and response_dict["ok"] == True:
        population = response_dict["body"]["population"]
        df_happiness_with_population.at[idx, "population"] = population
    else:
        print(f"Error fetching population for {country}")

# Print the updated dataframe
print(df_happiness_with_population)
```

```python
import http.client
from bs4 import BeautifulSoup
from tabulate import tabulate

conn = http.client.HTTPSConnection("world-population.p.rapidapi.com")

headers = {
    'X-RapidAPI-Key': "ff1ba047e1msh847b15bfd7c32e6p16051bjsne6e319c0a978",
    'X-RapidAPI-Host': "world-population.p.rapidapi.com"
}

conn.request("GET", "/worldpopulation", headers=headers)

res = conn.getresponse()
data = res.read().decode("utf-8")

# Parse the HTML response
soup = BeautifulSoup(data, 'html.parser')

# Find the table element
table = soup.find('table')

if table:
    # Extract table headers
    headers = [header.text.strip() for header in table.find_all('th')]

    # Extract table rows
    rows = []
    for row in table.find_all('tr'):
        row_data = [cell.text.strip() for cell in row.find_all('td')]
        if row_data:
            rows.append(row_data)

    # Print the table
    print(tabulate(rows, headers=headers))
else:
    print("No table found in the HTML response.")
```
